**Goal:** Uncover trends and patterns in Netflix's content library

# 1. Importing Libraries & Loading Data

```python
# Importing required libraries for data analysis
import numpy as np
import pandas as pd
```

```python
# Loading the Netflix dataset into a DataFrame
df=pd.read_csv('netflix_titles.csv')
```

# 3. Data Cleaning

```python
# Counting missing values in each column
df.isnull().sum()
```

```python
# Dropping rows where director is missing
df.dropna(subset=["director"],inplace=True)
```

```python
# Filling missing values in country, cast and rating columns
df['country'] = df['country'].fillna('Unknown')
df['cast']=df['cast'].fillna('No Cast')
df['rating']=df['rating'].fillna("No Rating")
df['duration']=df['duration'].fillna('Unknown')
```

 # 4. DateTime Handling

```python
# Converting date_added to datetime format and extracting year and month
df['date_added'] = pd.to_datetime(df['date_added'], format='mixed')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df[['date_added', 'year_added', 'month_added']].head()
```

# 5. Content Analysis

```python
# Counting Movies vs TV Shows
df['type'].value_counts()
```

```python
# Top 10 countries producing most content
df['country'].value_counts().head(10)
```

```python
# Content added to Netflix per year
df['year_added'].value_counts().sort_index()
```

```python
# Top 10 directors with most titles on Netflix
df['director'].value_counts().head(10)
```

# 6. String Operations

```python
# Splitting genres and finding top 10 most common genres
df['genre'] = df['listed_in'].str.split(',')
df_genres = df.explode('genre')
df_genres['genre'] = df_genres['genre'].str.strip()
```

# 7. GroupBy & Aggregation 

```python
# Extracting main genre from listed_in column
df_genres['genre'].value_counts().head(10)
df['main_genre'] = df['listed_in'].str.split(',').str[0].str.strip()
df['main_genre'].value_counts().head(10)
```

```python
# Content count by country and type for top 10 countries
top_countries = df['country'].value_counts().head(10).index
df[df['country'].isin(top_countries)].groupby(['country','type'])['title'].count()
```

```python
# Average release year by content type
df.groupby('type')['release_year'].mean()
```

```python
# Content added per year broken down by type
df.groupby(['year_added','type'])['title'].count()
```

```python
# Content count by rating category
df.groupby('rating')['title'].count().sort_values(ascending=False)
```

# 8. Pivot Tables

```python
# Pivot table showing content count by type and rating
pivot = pd.pivot_table(df, values='title', 
                       index='rating', 
                       columns='type', 
                       aggfunc='count')
pivot
```

```python
# Pivot table showing content added per year by type
pivot2 = pd.pivot_table(df, values='title',
                        index='year_added',
                        columns='type',
                        aggfunc='count')
pivot2
```

# 9. Interesting Insights

```python
# Which month has the most content added on Netflix
df['month_added'].value_counts().sort_index()
```

```python
# Total number of unique directors on Netflix
df['director'].nunique()
```

```python
# Top 10 most featured cast members on Netflix
df['cast'].str.split(',').explode().str.strip().value_counts().head(10)
```

```python
# Finding the 5 longest movies on Netflix
movies = df[df['type']=='Movie'].copy()
movies = movies[movies['duration'] != 'Unknown']
movies['duration_mins'] = movies['duration'].str.replace(' min','').astype(int)
movies.nlargest(5, 'duration_mins')[['title','duration_mins']]
```

## Key Findings
- United States dominates Netflix with the most content
- 2019 was the peak year for content added on Netflix
- International Movies is the most common genre
- Anupam Kher is the most featured actor on Netflix
- The longest movie on Netflix is The School of Mischief at 253 mins
